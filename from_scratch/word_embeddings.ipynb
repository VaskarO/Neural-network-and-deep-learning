## Word embeddings
#### Is the process of transforming words into vectors in a multi-dimensional space. The word are spread in the space in such a way that their distance corresponds to their semantic or lingutic similarity.

### Two main types
#### Frequency based embedding
#### Prediction based embedding

### Frequency based embedding
#### TF-IDF is a statistical measure used to evaluate the importance of a word to a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus.


```python
text=["the process of transforming words into vectors in a multi-dimensional space. The word are spread in the space in such a way that their distance corresponds to their semantic or lingutic similarity",
      "TF-IDF is a statistical measure used to evaluate the importance of a word to a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus"]
import pandas as pd
import numpy as np 
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer()
word_count=count.fit_transform(text)
print(word_count)
```

```python
print(word_count.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(word_count)
df_idf = pd.DataFrame(tfidf_transformer.idf_, columns=["idf_weights"])

df_idf.sort_values(by=['idf_weights'])
```
